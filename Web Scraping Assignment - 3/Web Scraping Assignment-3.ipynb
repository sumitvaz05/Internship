```python
driver = webdriver.Chrome('chromedriver.exe')

driver.get('https://www.google.com/maps')

search = driver.find_element_by_xpath("//div[@id= 'gs_lc50']/input")
search.send_keys('Mumbai, Maharashtra')

search_btn = driver.find_element_by_xpath("//div[@class= 'xoLGzf-BIqFsb-haAclf']/button")
search_btn.click()
```

```python
# Note the URl itself contains the coordinates(upto 7 decimal places) and their directions(N,E are positive S,W are negative)
url=driver.current_url

# Knowing that there are 7 decimal
coord = url[url.find("@")+1:url.find("@")+25]
coord
```

```python
# You can have the coordinates string be as long as 25 characters depending on the value so we narrow it down to second ooccurence of the ',' 
# The following program returns 2nd occurence of substring in string
def sec_occur(strng, subs):
    val=-1
    for i in range(0,2):
        val= strng.find(subs,val+1 )
    return val
```

```python
# We shorten the string to only contain the coordinates and a single ','
coord= coord[:sec_occur(coord, ',')]

# Splitting the string to two floating values
list_coord = coord.split(',')
list_coord

# We use the condition of positive/negative to get direction
if float(list_coord[0])>0:
    first_direction='North'
else: 
    first_direction='South'
    
if float(list_coord[1])>0:
    second_direction='East'
else:
    second_direction='West'

print(f'The coordinates are: {abs(float(list_coord[0]))} {first_direction} and {abs(float(list_coord[1]))} {second_direction}')
```
