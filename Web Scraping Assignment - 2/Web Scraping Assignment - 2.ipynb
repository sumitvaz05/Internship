```python
import time
import pandas as pd

import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
# Let's create some functions to make it easier

# First when there are no parent/child tags involved
def scrape_no_parent(lst, tag, clss, keyword):
    holder = driver.find_elements_by_xpath(f"//{tag}[@{clss}= '{keyword}']")

    for i in holder:
        lst.append(i.text)

# Parent/Child tags
def scrape_parent(lst, tag, clss, keyword, child):
    holder = driver.find_elements_by_xpath(f"//{tag}[@{clss}= '{keyword}']/{child}")

    for i in holder:
        lst.append(i.text)

# Creating dataframes with index from 1
def dataframe_creating(**kwargs):
    df_name = pd.DataFrame({})
    for key, value in kwargs.items():
        df_name[key] = value
    
    # Correcting the index    
    df_name.index +=1    
    
    return df_name
        
```

### Problem 7: Scrape 100 reviews data from flipkart.com for iphone11 phone

```python
# Initialise required lists

rating_list=[]
review_summ_list=[]
full_review_list=[]
```

```python
driver = webdriver.Chrome('chromedriver.exe')

url = 'https://www.flipkart.com/apple-iphone-11-black-64-gb-includes-earpods-power-adapter/product-reviews/itm0f37c2240b217?pid=MOBFKCTSVZAXUHGR&lid=LSTMOBFKCTSVZAXUHGREPBFGI&marketplace=FLIPKART&page=1'

# Since there are 10 reviews per page, a total of 10 pages will be enough
for i in range(1,11):
    url=url[:-1]+str(i)
    
    driver.get(url)

    # Ratings
    holder = driver.find_elements_by_xpath("//div[@class= '_3LWZlK _1BLPMq' or @class= '_3LWZlK _1rdVr6 _1BLPMq']")

    for i in holder:
        rating_list.append(i.text)
        
    # Review
    scrape_no_parent(review_summ_list, 'p', 'class', '_2-N8zT')

    # Full Review
    holder = driver.find_elements_by_xpath("//div[@class= 't-ZTKy']/div/div")

    for i in holder:
        full_review_list.append(i.text)
```

### Problem 9: Myntra Shoes (black, 6649-13099)

```python
driver = webdriver.Chrome('chromedriver.exe')

driver.get('https://www.myntra.com/shoes')
```

```python
# Let us select the filters

# For price range
# parent:: goes back to the parent node and from there we can maneuver to div, which is the click button
search_price = driver.find_element_by_xpath("//input[@value= '6908.0 TO 13602.0']/parent::*/div[1]")
search_price.click()
```

```python
# For colour of shoes
search_colour = driver.find_element_by_xpath("//span[@class= 'colour-label colour-colorDisplay']")
search_colour.click()
```

```python
# Initialise required lists

brand_list=[]
short_desc_list=[]
price_list=[]
```

```python
url='https://www.myntra.com/shoes?f=Color%3ABlack_36454f&rf=Price%3A6908.0_13602.0_6908.0%20TO%2013602.0'
driver.get(url)
    
# Brand
scrape_no_parent(brand_list, 'h3', 'class', 'product-brand')

# Short Description
scrape_no_parent(short_desc_list, 'h4', 'class', 'product-product')

# Price
scrape_no_parent(price_list, 'span', 'class', 'product-discountedPrice')

# Now we shall click on the 'Next' button to go to page 2 and do it again
next_page = driver.find_element_by_xpath("//li[@class= 'pagination-next']/a[1]")

# Brand
scrape_no_parent(brand_list, 'h3', 'class', 'product-brand')

# Short Description
scrape_no_parent(short_desc_list, 'h4', 'class', 'product-product')

# Price
scrape_no_parent(price_list, 'div', 'class', 'product-price') 
```

```python
# Some prices have stuck together, leading to a smaller number of entries in price_list eg 'Rs. 7999Rs. 9999(20% OFF)'
#Let us define a function which finds a second occurence of 'R' and split the entry into 2 
def find_2nd(string, substring):
    return string.find(substring, string.find(substring) + 1)

# Create a new list 
new_price=[]
for i in price_list:
    c=find_2nd(i, 'R')
    
    # The reasoning here is only entries that are stuck together will be higher than 9 characters so we can safely split them 
    # and just simply append the rest
    if len(i)>10:
        new_price.append(i[:c])
        new_price.append(i[c:])
    else:
        new_price.append(i)
    
len(new_price)
        
```

```python
kwargs={'Brand-Name':brand_list, 'Product Description':short_desc_list, 'Price': new_price}

df = dataframe_creating(**kwargs)

df
```
