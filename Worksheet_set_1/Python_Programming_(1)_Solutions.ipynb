```python
# WAP to find factorial of a number
def factorial():
    try:
        n=int(input('Provide an integer: '))
        if n>1:
            return (n*recursive(n-1))
        elif n==1:
            return 1
        else:
            print('Negative integers and zero not allowed. 0! is assumed as 1')
            
    except ValueError:
        print('Only positive integers allowed')


def recursive(n):
    if n>1:
        return n*recursive(n-1)
    elif n==1:
        return 1        

# Example    
factorial()
```

```python
# WAP to find whether a number is prime or composite
def is_prime(n):
    try:
        potential_factors=list(range(2, n//2+1))  # Potential factors can only be upto n/2
        factor_found=False                        # We'll use this at the end to confirm prime or not

        for i in potential_factors:
            if n%i==0:
                factor_found=True
                print('Composite')
                break
            else:
                continue
        if factor_found==False:
            print('Prime')
            
    except:
        print('Only positive integers allowed')
        
# Example
is_prime(10002563)
```

```python
# WAP to find whether string is palindrome or not
def is_palindrome(s):
    return s == s[::-1]

# Example
is_palindrome('level')
```
